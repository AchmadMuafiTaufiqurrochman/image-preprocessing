```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# Buka gambar asli
img = Image.open("images/mobil.jpeg")

# Buat salinan gambar untuk negatif
img_n = img.copy()

# Konversi gambar menjadi array NumPy
img_array = np.array(img_n)

# Transformasi gambar negatif
img_negatif = 255 - img_array  # Membalik warna

# Konversi kembali ke format gambar
img_n = Image.fromarray(img_negatif)

# Fungsi untuk menampilkan histogram
def plot_histogram(img, title, ax):
    # Pisahkan channel warna (R, G, B)
    r, g, b = img.split()
    
    # Hitung histogram untuk masing-masing channel
    ax.hist(np.array(r).ravel(), bins=256, color='red', alpha=0.6, label="Red")
    ax.hist(np.array(g).ravel(), bins=256, color='green', alpha=0.6, label="Green")
    ax.hist(np.array(b).ravel(), bins=256, color='blue', alpha=0.6, label="Blue")
    
    ax.set_title(title)
    ax.set_xlim(0, 255)
    ax.legend()

# Plot hasilnya
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Gambar asli
axes[0, 0].imshow(img)
axes[0, 0].set_title("Gambar Asli")
axes[0, 0].axis("off")

# Gambar negatif
axes[0, 1].imshow(img_n)
axes[0, 1].set_title("Gambar Negatif")
axes[0, 1].axis("off")

# Histogram gambar asli
plot_histogram(img, "Histogram Gambar Asli", axes[1, 0])

# Histogram gambar negatif
plot_histogram(img_n, "Histogram Gambar Negatif", axes[1, 1])

plt.tight_layout()
plt.show()
```
